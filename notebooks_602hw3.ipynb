```python
hours_worked = float(input("Enter the number of hours employee worked"))
hourly_pay = float(input("Enter the hourly pay rate"))
if hours_worked <= 20:
    regular = hours_worked * hourly_pay
    overtime = 0
else:
    overtime = hours_worked * 1.5 * hourly_pay
    regular = 20 * hourly_pay
    hours_worked_over = hours_worked - 20
gross = regular + overtime
print("hours worked", hours_worked)
print("overtime hours worked", hours_worked_over)
print("regular pay: ", regular)
print("overtime pay:", overtime)
print("gross pay:", gross)
```

```python
def q5():
    total = 0
    n = 30
    for i in range(1, n + 1):
        numerator = i 
        denominator = (n + 1) - i 

        total += numerator / denominator
        
    return total

result = q5()
print('the total sum of the series is', round(result))
```
