```python
days = ["monday","tuesday","wednesday","thursday","friday","saturday","sunday"]

sales = []

for day in days:
    sale = float(input("enter store sales for : " + str(day)))
    sales.append(sale)
            
total_sales = sum(sales)

print("the total sales of the week are $", total_sales)
    
        
```
